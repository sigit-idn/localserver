```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
import re
from PIL import Image
import os

options = Options()
options.binary_location = r'C:\Users\nobor\AppData\Local\Mozilla Firefox\firefox.exe'
service = Service(executable_path=r'geckodriver.exe')

driver = webdriver.Firefox(service=service, options=options)

product_numbers = ("202-0506", "202-0508", "202-0512", "202-0514", "202-0516", "202-0518", "202-0520", "202-0492", "202-0493", "202-0494", "202-0497", "202-0499", "202-0502", "202-0504")

for product_number in product_numbers:
    driver.get("https://www.rakuten.ne.jp/gold/tayu-tafu/iframe/shot/%s-shot.html" % product_number)
    driver.set_window_size(650, driver.execute_script("return document.body.scrollHeight"))

    dls = driver.find_elements(By.CSS_SELECTOR, "#pageShot dl")
    bs = driver.find_elements(By.CSS_SELECTOR, "#pageShot b")

    for dlIndex in range(len(dls)):
        dl = dls[dlIndex]
        b = bs[dlIndex].text
        dl.screenshot( product_number + "-sp-%s.png" % re.sub(r'-0*', '-t', b).lower())
        im = Image.open(product_number + "-sp-%s.png" % re.sub(r'-0*', '-t', b).lower())
        rgb_im = im.convert('RGB')
        rgb_im.save(product_number + "-sp-%s.jpg" % re.sub(r'-0*', '-t', b).lower())
        os.remove(product_number + "-sp-%s.png" % re.sub(r'-0*', '-t', b).lower())

    time.sleep(1)

driver.quit()
```
