```python
# start selenium with installing chromedriver
import time
import json
from selenium import webdriver
from selenium.webdriver.common.by import By

from selenium.webdriver.firefox.service import Service

from selenium.webdriver.firefox.options import Options
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary


options = Options()
options.binary_location = r'C:\Users\nobor\AppData\Local\Mozilla Firefox\firefox.exe'
service = Service(executable_path=r'geckodriver.exe')

driver = webdriver.Firefox(service=service, options=options)
driver.delete_all_cookies()

product_numbers = (
  "202-0506",
  "202-0508",
  "202-0512",
  "202-0514",
  "202-0516",
  "202-0518",
  "202-0520",
  "202-0492",
  "202-0493",
  "202-0494",
  "202-0497",
  "202-0499",
  "202-0502",
  "202-0504",
)

iframe_height_data = {}


for product_number in product_numbers:
    driver.get("https://www.rakuten.ne.jp/gold/tayu-tafu/iframe/shot/%s-shot.html" % product_number)
    driver.set_window_size(375, driver.execute_script("return document.body.scrollHeight"))
    shot_sp_height = int(driver.find_element(By.TAG_NAME, "body").get_attribute("scrollHeight")) - 600
    
    driver.set_window_size(650, driver.execute_script("return document.body.scrollHeight"))
    shot_pc_height = int(driver.find_element(By.TAG_NAME, "body").get_attribute("scrollHeight"))

    driver.get("https://www.rakuten.ne.jp/gold/tayu-tafu/iframe/colors/%s-c.html" % product_number)
    driver.execute_script("document.querySelector('head').innerHTML += '<meta name=\"viewport\" content=\"width=device-width, initial-scale=1.0, maximum-scale=1.0, user-scalable=0\">'")
    driver.execute_script("document.querySelector('body > div').removeAttribute('style')")
    driver.set_window_size(375, driver.execute_script("return document.body.scrollHeight"))
    colors_sp_height = int(driver.find_element(By.ID, "colors").size["height"] / 1.17)

    driver.set_window_size(650, driver.execute_script("return document.body.scrollHeight"))
    driver.execute_script("document.querySelector('head').innerHTML += '<meta name=\"viewport\" content=\"width=device-width, initial-scale=1.0, maximum-scale=1.0, user-scalable=0\">'")
    driver.execute_script("document.querySelector('body > div').removeAttribute('style')")
    colors_pc_height = int(driver.find_element(By.ID, "colors").size["height"] / 1.1 + 100)

    
    driver.get("https://www.rakuten.ne.jp/gold/tayu-tafu/iframe/spec/%s-s.html" % product_number)
    driver.set_window_size(375, driver.execute_script("return document.body.scrollHeight"))
    spec_sp_height = int(driver.find_element(By.TAG_NAME, "body").get_attribute("scrollHeight"))

    driver.set_window_size(650, driver.execute_script("return document.body.scrollHeight"))
    spec_pc_height = int(driver.find_element(By.TAG_NAME, "body").get_attribute("scrollHeight"))

    iframe_height_data[product_number] = '''
#item-review iframe {
  height: %spx;
  }

@media screen and (max-width: 767px) {
  #item-review iframe {
    height: calc(%spx - 3.75px * 120 + 120vw);
    }

  #spec iframe {
    height: %spx;
    }
}
    ''' % (colors_pc_height, colors_sp_height, spec_sp_height)


driver.quit()

with open("iframe_heights_data.json", "w") as f:
  f.write(
    json.dumps(iframe_height_data, indent=2, ensure_ascii=False)
  )


print("done")



```
