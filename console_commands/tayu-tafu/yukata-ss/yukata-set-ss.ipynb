```python
# start selenium with installing chromedriver
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

from selenium.webdriver.firefox.service import Service

from selenium.webdriver.firefox.options import Options
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary


options = Options()
options.binary_location = r'C:\Users\nobor\AppData\Local\Mozilla Firefox\firefox.exe'
service = Service(executable_path=r'geckodriver.exe')

driver = webdriver.Firefox(service=service, options=options)

product_numbers = ("202-0508", "202-0520", "202-0494", "202-0497", "202-0502")

for product_number in product_numbers:
    driver.get("https://www.rakuten.ne.jp/gold/tayu-tafu/iframe/shot/%s-shot.html" % product_number)
    driver.set_window_size(650, driver.execute_script("return document.body.scrollHeight"))

    pageShot = driver.find_element(By.ID, "pageShot")
    # pageShot.screenshot( product_number + "-sp-pt.jpg")

    with open("group-t.js", "r") as f:
        script = f.read()
        driver.execute_script(script)

    t_numbers = [i for i in range(1, 50, 2)]

    t_groups = driver.find_elements(By.CLASS_NAME, "t-group")
    for t_index in range(len(t_groups)):
        t_group = t_groups[t_index]

        t_number = t_numbers[t_index]
        t_group.screenshot( product_number + "-sp-t%d-%d.jpg" % (t_number, t_number + 1))

    time.sleep(3)
    
    driver.get("https://www.rakuten.ne.jp/gold/tayu-tafu/iframe/spec/%s-s.html" % product_number)
    driver.execute_script("document.body.style.padding = '1px'")
    driver.find_element(By.TAG_NAME, "body").screenshot( product_number + "-sp-s.jpg")

    time.sleep(1)

    driver.get("https://www.rakuten.ne.jp/gold/tayu-tafu/iframe/colors/%s-c.html" % product_number)
    driver.execute_script("document.body.style.transform = 'scale(0.8)'")
    colorsWrapper = driver.find_element(By.ID, "colors")
    colorsWrapper.screenshot( product_number + "-sp-c.jpg")
    colors = colorsWrapper.find_elements(By.CSS_SELECTOR, "ul li")

    image_title_index = 1
    range_values = ((1, 6),(7, 12), (13, len(colors) - 12))

    for range_value in range_values:
        start, end = range_value
        for color_index in range(len(colors)):
            if color_index < start or color_index > end:
                colors[color_index].execute_script("this.style.display = 'none'")
            else:
                colors[color_index].execute_script("this.style.display = 'block'")
                
            colorsWrapper.screenshot( product_number + "-sp-c_%d.jpg" % image_title_index)
            image_title_index += 1

    time.sleep(2)

```

```python
# start selenium with installing chromedriver
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

from selenium.webdriver.firefox.service import Service

from selenium.webdriver.firefox.options import Options
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary


options = Options()
options.binary_location = r'C:\Users\nobor\AppData\Local\Mozilla Firefox\firefox.exe'
service = Service(executable_path=r'geckodriver.exe')

driver = webdriver.Firefox(service=service, options=options)

product_numbers = (
                    "202-0506",
                    "202-0508",
                    "202-0512",
                    "202-0514",
                    "202-0516",
                    "202-0518",
                    "202-0520",
                    "202-0492",
                    "202-0493",
                    "202-0494",
                    "202-0497",
                    "202-0499",
                    "202-0502",
                    "202-0504",
                    )

iframe_height_data = {}

for product_number in product_numbers:
    driver.get("https://www.rakuten.ne.jp/gold/tayu-tafu/iframe/shot/%s-shot.html" % product_number)
    driver.set_window_size(375, driver.execute_script("return document.body.scrollHeight"))
    iframe_height_data[product_number] = {}
    iframe_height_data[product_number]["shot"] = {}
    iframe_height_data[product_number]["shot"]["sp"] = driver.execute_script("return document.body.scrollHeight")

    driver.set_window_size(650, driver.execute_script("return document.body.scrollHeight"))
    iframe_height_data[product_number]["shot"]["pc"] = driver.execute_script("return document.body.scrollHeight")

    driver.get("https://www.rakuten.ne.jp/gold/tayu-tafu/iframe/colors/%s-c.html" % product_number)
    driver.set_window_size(375, driver.execute_script("return document.body.scrollHeight"))
    iframe_height_data[product_number] = {}
    iframe_height_data[product_number]["colors"] = {}
    iframe_height_data[product_number]["colors"]["sp"] = driver.execute_script("return document.body.scrollHeight")


    driver.set_window_size(650, driver.execute_script("return document.body.scrollHeight"))
    iframe_height_data[product_number]["colors"]["pc"] = driver.execute_script("return document.body.scrollHeight")
    
    driver.get("https://www.rakuten.ne.jp/gold/tayu-tafu/iframe/spec/%s-s.html" % product_number)
    driver.set_window_size(375, driver.execute_script("return document.body.scrollHeight"))
    iframe_height_data[product_number] = {}
    iframe_height_data[product_number]["spec"] = {}
    iframe_height_data[product_number]["spec"]["sp"] = driver.execute_script("return document.body.scrollHeight")

    driver.set_window_size(650, driver.execute_script("return document.body.scrollHeight"))
    iframe_height_data[product_number]["spec"]["pc"] = driver.execute_script("return document.body.scrollHeight")

    driver.quit()

import json
with open("iframe_height_data.json", "w") as f:
    json.dump(iframe_height_data, f)


```
